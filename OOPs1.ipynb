```python
class Atm:

  # constructor -- a function inside the class

  # constructor is a  special fn : -> superpower -> no need to call constructor explicitly to run code in it.
  #  as we make a object of it.all code writtennin it will run automatically


  def __init__(self):
    print(id(self)) # address of self
    self.pin =" "
    self.balance =0
     # self.menu()


  def menu(self):
    user_input = input("""
    Hi how can i help u?
    1. press 1 to creat pin
    2. press 2 to change pin
    3. press 3 to check balance
    4. press 4 to withdraw
    5. anything else to exit
    """)

    if user_input == '1':
      self.create_pin()
    elif user_input == "2":
      # change pin
      self.change_pin()
    elif user_input == "3":
      # check balance
      self.check_balance()
    elif user_input == "4":
      # withdraw
      self.withdraw()
    else:
      exit()



  def create_pin(self):
    user_pin = input("enter your pin")
    self.pin = user_pin
    print("pin set successfully")

    user_balance=int(input('enter balance'))
    self.balance = user_balance
    print("balance added successfully")
    self.menu()

  def change_pin(self):
    old_pin=input('enter old pin')

    if old_pin == self.pin:
      # let him chnage the pin
      new_pin = input("enter new pin")
      self.pin = new_pin
      print("pin change successfully")
      self.menu()

    else:
      print("bhaag jao")
      self.menu()


  def check_balance(self):
    user_pin=input("enter your pin")
    if user_pin == self.pin:
      print("ur balance is : ",self.balance)
    else:
      print("bhaag jao")


  def withdraw(self):
    user_pin=input("Enter ur pin")
    if user_pin == self.pin:
      # allow withdraw
      amount =int(input("enter the amt"))
      if amount <= self.balance:
        self.balance-=amount
        print("withdraw success. balance is 5",self.balance)
      else:
        print("mat kar lala")
    else:
      print(" sale chor")
    self.menu()













```

```python
class Fraction:

  # parameterised constructor
  def __init__(self,x,y):
    self.num=x
    self.den=y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,other):
    new_num=self.num*other.den+other.num*self.den
    new_den=self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __sub__(self,other):
    new_num=self.num*other.den-other.num*self.den
    new_den=self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __multi__(self,other):
    new_num=self.num*other.num
    new_den=self.den*other.den
    return '{}/{}'.format(new_num,new_den)

  def __truediv__(self,other):
    new_num=self.num*other.den
    new_den=self.den*other.num
    return '{}/{}'.format(new_num,new_den)


  def convert_to_decimal(self):
    return self.num/self.den


```

```python
fr1=Fraction(4,9)
```

```python
fr2=Fraction(3,4)
```

```python
print(fr1+fr2)
```

```python
fr1.convert_to_decimal()
```
